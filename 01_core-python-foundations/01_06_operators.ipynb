# 1. Arithmetic Operators ( +, -, *, /, //, %, ** )

An **operator** is a symbol that tells Python to do an action on values.

Example in real life:

- `2 + 3` → “add 2 and 3”

- `10 * 2` → “multiply 10 and 2”

In this section, we learn arithmetic operators (math operators):

- `+` add

- `-` subtract

- `*` multiply

- `/` divide (always gives a decimal answer in Python)

- `//` floor division (gives the whole number part only)

- `%` modulo (gives the remainder)

- `**` power (exponent)

Common confusion (important)

- `6 / 3` gives `2.0` (a decimal)

- `6 // 3` gives `2` (whole number)

Another common confusion: order matters

Python follows operator precedence:

Multiplication happens before addition
- `2 + 3 * 4` → multiply first → `2 + 12` → `14`

Parentheses () can change the order
`(2 + 3) * 4` → add first → `5 * 4` → `20`

```python
a = 10
b = 3
```

```python
print(a + b)
print(a - b)
print(a * b)
```

```python
print(a / b)     # normal division (decimal)
print(a // b)   # floor division (whole number part)
print(a % b)     # remainder
```

```python
print(b ** 2)   # power
```

```python
print(2 + 3 * 4) 
print((2 + 3) * 4)
```

```python
print(6 / 3)
print(6 // 3)
```

# 2. Comparison Operators ( ==, !=, >, <, >=, <= )

**Comparison operators** are used to **compare two values**.

The result of a comparison is always a Boolean:

`True` (means: yes, it’s correct)

`False` (means: no, it’s not correct)

**Comparison operators you will use a lot**

- `==` equal to

- `!=` not equal to

- `>` greater than

- `<` less than

- `>=` greater than or equal to

- `<=` less than or equal to

**Very common mistake (super important)**

- `=` means assignment (store a value in a variable)

- `==` means comparison (check if two values are equal)

Example:

- `x = 5` → put 5 into x

- `x == 5` → “is x equal to 5?”

```python
x = 10
y = 7
```

```python
print(x == y)
print(x != y)
```

```python

print(x > y)
print(x < y)
```

```python

print(x >= 10)
print(y <= 7)

```

```python
# Comparison with numbers directly
print(5 == 5)
print(5 != 6)
print(3 > 9)
```

# 3. Logical Operators ( and, or, not )

**Logical operators** are used to **combine** or **reverse** Boolean values `(True / False)`.

Think of it like decision-making in real life:

- **and** → both conditions must be True

    Example: “I will go out if it is Saturday AND I am free.”

- **or** → at least one condition must be True

    Example: “I will order food if it is rainy OR I am tired.”

- **not** → flips the result

    Example: “not True” becomes False

**1) and**

- `True and True` → `True`

    - If any side is False → result is False

**2) or**

- `False or True` → `True`

    - If both are False → result is False

**3) not**

- `not True` → `False`

- `not False` → `True`

- `True` becomes `False`, and `False` becomes `True`

```python
x = 10
```

```python

print((x > 5) and (x < 20))
print((x > 5) and (x < 8))
```

```python

print((x < 5) or (x == 10))
print((x < 5) or (x == 7))
```

```python

print(not (x == 10))
print(not (x > 100))
```

```python

# Direct boolean examples
print(True and False)
print(True or False)
print(not True)
```

# 4. Assignment Operators ( =, +=, -=, *=, /=, //=, %=, **= )

**Assignment operators** are used to **store** a value in a variable, and sometimes **update** it in the same step.

1) Basic assignment

- `=` means: “put this value inside the variable”

        - `x = 10` → now `x` holds `10`

2) **Update-and-assign (shortcut)**

These are shortcuts for “take the old value, do something, store it back”.

Example:

- `x += 5` means: `x = x + 5`

- `x -= 2` means: `x = x - 2`

- `x *= 3` means: `x = x * 3`

- `x /= 2` means: `x = x / 2` (becomes float)

- `x //= 2` means: `x = x // 2` (floor division)

- `x %= 3` means: `x = x % 3`

- `x **= 2` means: `x = x ** 2`

**Common confusion (important)**

- `x = x + 1` is totally valid, but `x += 1` is shorter and very common.

- `/=` usually creates a **decimal** value (float), even if numbers look “whole”.

```python
x = 10
print(x)
```

```python

x += 5
print(x)
```

```python

x -= 3
print(x)
```

```python

x *= 2
print(x)
```

```python

x /= 4
print(x)
```

```python

y = 17
print(y)
```

```python

y //= 3
print(y)
```

```python

y %= 4
print(y)
```

```python

z = 3
print(z)
```

```python

z **= 2
print(z)
```
