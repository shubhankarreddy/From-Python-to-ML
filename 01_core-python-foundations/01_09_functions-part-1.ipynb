# **Functions**


A **function** is a small reusable block of code.


**Use Case**:

1. Avoid repeating code

    If you do the same thing many times, write it once in a function.

2. Make code clean

    Your main code becomes easier to read.

3. Easy to update

    Change logic in one place, it updates everywhere you call the function.

syntax : 

def function_name():
    # code inside function

- `def` is a keyword that means “define a function”

- Function name should be simple (like `greet`, `add`, `check_even`)

- `()` are required

- `:` is required

- Indentation is required (same rule as if/loops)

```python
# Example

def greet():
    print("Hello")
```

```python
# Calling function

greet()
```

## **Functions with inputs**

A function can take input values.
These inputs are called parameters.

**Key words**

- **Parameter** : the variable written inside `()` in the function definition
    Example: `def greet(name)`: → `name` is a parameter

- **Argument** : the actual value you pass when calling the function
    Example: `greet("Amit")` → `"Amit"` is an argument

syntax :

def function_name(parameter):
    # use parameter

```python
# Example : Creating function by the name greet.

def greet(name):
    print("Hello", name)
```

```python
# Calling function

greet("Amit")
```

```python
# Creating function double_num that multiplies given number by 2.

def double_num(x):
    x *= 2
    print(x)
```

```python
# Calling function

double_num(5)
```

## `function` with `return` statement

Till now our functions printed something.

But many times we want a function to give back a value so we can use it later.

That is what `return` does.


**`print` vs `return` (important)**

- `print()` → shows output on the screen (only for us to see)

- `return` → sends a value back to where the function was called

If you want to **store the result in a variable**, you need `return`.

```python
# Creating function add to add two numbers.

def add(a, b):
    result = a + b
    return result
```

```python
# Calling function add

x = add(10, 5)
print(x)
```

**Explaination of above add function:**

1. We ask the function to add 10 and 5.

2. The 'return' command takes the answer (15) and hands it back to us out of the function.

3. Because we wrote 'x =', that answer is saved into the variable 'x'.

Simply put: `return` is how a function gives you its final answer so you can actually use it.


*Now if we used `print` instead of `return`, it would just show the answer on the screen, but it wouldn't hand anything back to save. So, 'x' would be totally empty (None).*

```python
# Example 2 : If the bill amount is more than or equal to 1000 then give a 100 rs discount.

def final_bill(bill):
    if bill >= 1000:
        bill -= 100
    return bill

# After calculation return gave the final output back to final_bill.
```

```python
# Running bill function with 1200 as bill amount

amount = final_bill(1200)
print(amount)

# amount variable is saving the output of final_bill function.
```

```python
# Bill function with 700 as bill amount

amount = final_bill(700)
print(amount)
```
